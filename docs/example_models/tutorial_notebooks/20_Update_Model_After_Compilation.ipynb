# Model update

Since model compilation can sometimes take an extensive amount of time for some models, we introduced the update model feature. This feature allows the user to change a parameter or species concentration from the model after it has been compiled. In this section, we provide an example.


```python
from mobspy import *

A = BaseSpecies()
k_a = ModelParameters(0)

A >> Zero [k_a]

S = Simulation(A)
S.duration = 10
S.run()
```

In the previous model, we forgot to add an initial count to the A species and unfortunately set the parameter k_a to zero. 
The method update_model from the simulation object allows us to correct this. 
For each value to be updated, the user must supply a two-element list to the update model function. This two-element list contains in the first index either the meta-species whose initial count will be updated or the parameter whose value will be updated, the second index is the new updated value. Bellow an example follows: 

```python
S.update_model([A, 100], [k_a, 1])
S.run()
```

Similarly to the initial assignments, the user can also update meta-species initial counts using queries. 
Furthermore, the meta-species count update follows the same rules as an initial count setting, with the default state being taken when no query is specified. It is also compatible with the All[] operator.
